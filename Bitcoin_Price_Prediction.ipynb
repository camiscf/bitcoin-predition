## Analisar por a média de open e close por mês

```python
bitcoin_df['Date']=pd.to_datetime(bitcoin_df['Date'], format='%Y-%m-%d')
```

```python
monthvise= bitcoin_df.groupby(bitcoin_df['Date'].dt.strftime('%B'))[['Open','Close']].mean()
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
             'September', 'October', 'November', 'December']
monthvise = monthvise.reindex(new_order, axis=0)
monthvise
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=monthvise.index,
    y=monthvise['Open'],
    name='Stock Open Price',
    marker_color='crimson'
))
fig.add_trace(go.Bar(
    x=monthvise.index,
    y=monthvise['Close'],
    name='Stock Close Price',
    marker_color='lightsalmon'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45,
                  title='Monthwise comparision between Stock open and close price')
fig.show()
```

```python
month_high_low= bitcoin_df.groupby(bitcoin_df['Date'].dt.strftime('%B'))[['High','Low']].mean()
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
             'September', 'October', 'November', 'December']
month_high_low = month_high_low.reindex(new_order, axis=0)
month_high_low
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=month_high_low.index,
    y=month_high_low['High'],
    name='Stock High Price',
    marker_color='lightblue'
))
fig.add_trace(go.Bar(
    x=month_high_low.index,
    y=month_high_low['Low'],
    name='Stock Low Price',
    marker_color='lightgreen'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45,
                  title='Monthwise comparision between Stock high and low price')
fig.show()
```

### Maneiras mais simples de previsão
A primeira forma é usar a média dos ultimos 3 meses para prever os proximos 3 meses


```python
bitcoin_df['Return'] = bitcoin_df['Close'] - bitcoin_df['Open']
```

```python
bitcoin_df.dropna()
```

```python
mean_return_by_month = bitcoin_df.groupby(bitcoin_df['Date'].dt.strftime('%B'))[['Return']].mean()
mean_return_by_month = mean_return_by_month.reindex(new_order, axis=0)
```

```python
mean_return_by_month
```

```python
last_three_months = mean_return_by_month.iloc[-3:]
last_three_months
```

```python
next_three_months = {'October':'Month 1','November':'Month 2','December':'Month 3'}
last_three_months= last_three_months.rename(index = next_three_months)
```

```python
last_three_months
```

```python
fig = go.Figure()
# Create and style traces
fig.add_trace(go.Scatter(x=last_three_months.index, y=last_three_months['Return'], name='Return',
                         line=dict(color='firebrick', width=4)))

# Edit the layout
fig.update_layout(title='Relation between',xaxis_title='Month')


fig.show()
```

Agora vamos para uma técnica um pouco mais avançada que seria a regressão linear, nossa tarefa é prever o retorno da bitcoin (y) com base em variáveis (X), tais como open, close, high, low, volume.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
bitcoin_df = bitcoin_df.dropna()
bitcoin_df
```

```python
y = bitcoin_df[['Return']]
y
```

```python
X = bitcoin_df[['Open','Close','Volume','High','Low','Return']]
X
```

https://dadosaocubo.com/regressao-com-scikit-learn/

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
# Criando o modelo LinearRegression
regr = LinearRegression()
# Realizar treinamento do modelo
regr.fit(X_train, y_train)
```

```python
# Obtain the coefficient of determination by calling the model with the score() function, then print the coefficient:
r_sq = regr.score(X_test, y_test)
print('coefficient of determination:', r_sq)
# Print the Intercept:
print('intercept:', regr.intercept_)
# Print the Slope:
print('slope:', regr.coef_)
# Predict a Response and print it:
y_pred = regr.predict(X_test)
print('Predicted response:', y_pred, sep='\n')
```
